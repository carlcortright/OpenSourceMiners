```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter, OrderedDict
import collections

%matplotlib inline
sns.set()

projectsData = pd.read_csv("/Users/shubha/Documents/OpenSourceMiners/code/small_data/projects-1.2.0-2018-03-12.csv", low_memory=False)
#projectsData.info()
#projectsData.columns = ['Latest Release Publish Timestamp', 'Language']

d3 = projectsData.loc[projectsData['Language'] == 'JavaScript']
d3 = d3.dropna(subset=['Latest Release Publish Timestamp'])

#print(d3.info())
#print(d3)
languageVals = d3['Language']


#print(len(languageVals))
dateVals = pd.to_datetime(d3['Latest Release Publish Timestamp']).dt.date

dateList = []
for x in dateVals:
    dateList.append(x.strftime('%Y-%m'))


dateList2 = sorted(dateList, key=lambda d: tuple(map(int, d.split('-'))))
#print(dateList2)


class OrderedCounter(Counter, OrderedDict):
    'Counter that remembers the order elements are first encountered'

    def __repr__(self):
        return '%s(%r)' % (self.__class__.__name__, OrderedDict(self))

    def __reduce__(self):
        return self.__class__, (OrderedDict(self),)


c = OrderedCounter(dateList2)


dateV = []
dateF = []

for key, value in c.items():
    dateV.append(key)
    dateF.append(value)
    
plt.plot(dateV, dateF, 'ro', color='mediumslateblue')
plt.xlabel('Date')
plt.ylabel('Contributions')
plt.title('Contributions with JavaScript Over Time')
frequency = 15
plt.xticks(dateV[::frequency], dateV[::frequency])
plt.show()

#g = dateVals.groupby(pd.TimeGrouper("M"))
#print(g)


#dateClean = data.dropna(subset=['Age'])



b = projectsReposData['Dependent Projects Count']
#print(b)


```
